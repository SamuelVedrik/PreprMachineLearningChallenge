```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
URL = 'https://www23.statcan.gc.ca/imdb/p3VD.pl?Function=getVD&TVD=380372&CVD=380373&CLV=0&MLV=5&D=1'
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
headers =[header.a for header in soup.find_all(class_ = "bg-def-1")]
headers = [header.contents[0] for header in headers]
```

```python
descriptions = [par.prettify() for par in soup.find_all("p")][2:]
descriptions = [re.sub("|<.*>", "", desc) for desc in descriptions]
descriptions = [desc for desc in descriptions if desc != "\n \n  Exclusion(s)\n \n\n"]
```

```python
descriptions
```

```python
headers
```

```python
desc_dict = {k: v for k, v in zip(headers, descriptions)}
```

```python
df = pd.DataFrame.from_dict(desc_dict, orient="index").reset_index().rename(columns=
                                                                            {"index":"Name", 0:"Description"})

df.to_csv("NAICS_desc.csv")
```
